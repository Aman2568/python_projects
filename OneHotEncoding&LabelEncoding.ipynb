```python
import pandas as pd
```

```python
df = pd.read_csv("bmi_data.csv")
```

```python
df
```

```python
df.head(20)
```

```python
x = df[["Height_cm", "Weight_kg"]]
```

```python
y = df.Gender
```

```python
x
```

```python
y
```

One Hot Encoding

```python
a = pd.get_dummies(y)
```

```python
a
```

```python
from sklearn.tree import DecisionTreeClassifier    # DecisionTreeClassifier from sklearn.tree as OneHotEncoding works with DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x, a)
```

```python
ans = model.predict([[200,100]])
```

```python
ans
```

```python
if ans[0][0] == True:
  print("Female")
else:
  print("Male")
```

Lable Encoding

```python
from sklearn.preprocessing import LabelEncoder   #LabelEncoder from SKlearn.preprocessing for Label Encoding
```

```python
le = LabelEncoder()
```

```python
b = le.fit_transform(y)
```

```python
model1 = DecisionTreeClassifier()
```

```python
model1.fit(x, b)
```
